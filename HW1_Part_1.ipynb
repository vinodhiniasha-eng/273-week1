```python
## Your code for LP 1 goes here.
df = df.replace("?", np.nan)
df.isna().sum()
```

```python
# Column indices from the prompt
cat_missing_cols = [0, 3, 4, 5, 6]   # A1, A4, A5, A6, A7
num_missing_cols = [1, 13]           # A2, A14

# Fill categorical with mode
for col in cat_missing_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Ensure numeric types for continuous columns, then fill with median
for col in num_missing_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df[col].fillna(df[col].median())

df.isna().sum()[[0,1,3,4,5,6,13]]
```
